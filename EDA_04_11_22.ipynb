```python
import seaborn as sns

for column in ['alpha', 'delta', 'u', 'g', 'r', 'i', 'z', 'redshift']:
    plt.figure(figsize=(12,6))
    sns.histplot(data=df, x=column, kde=True, hue='class')
    plt.title(column)
    plt.show()
```

```python
import seaborn as sns

for column in ['alpha','delta','u', 'g', 'r', 'i', 'z', 'redshift']:
    plt.figure(figsize=(12,6))
    sns.boxplot(data=df, x=column, y='class', palette="flare", linewidth=1.5, saturation=0.6)
    plt.title(column)
    plt.show()
```
